```python
import os
import re
import pandas as pd
```

```python
#### TESTS
param_type = 'AP' 

folder = 'D:/OneDrive - Universidad Complutense de Madrid (UCM)/Doctorado/Curriculum_Learning/Easy_Multidigit_Addition_Decimal/'
folder_specific = f'{folder}v1-Exponential_Decay/'

analyzed_data_folder = f'{folder_specific}Analyzed_Data'

folder_path = f'{folder_specific}Tests'

output_path = f'{folder_specific}tests_results_addition_decimal.xlsx'

total_count = 5050
count_train = 4858
total_count_carry_over = 2025
count_carry_over_train = 1929
total_count_small = 820
count_small_train = 724
total_count_carry_over_small = 270
count_carry_over_small_train = 222
total_count_large = 3159
count_large_train = 3063
total_count_carry_over_large = 1296
count_carry_over_large_train = 1248

count_test = total_count - count_train

count_carry_over_test = total_count_carry_over - count_carry_over_train
total_count_no_carry_over = total_count - total_count_carry_over
count_no_carry_over_train = count_train - count_carry_over_train
count_no_carry_over_test = count_test - count_carry_over_test

count_small_test = total_count_small - count_small_train
count_large_test = total_count_large - count_large_train

count_carry_over_small_test = total_count_carry_over_small - count_carry_over_small_train
total_count_no_carry_over_small = total_count_small - total_count_carry_over_small
count_no_carry_over_small_train = count_small_train - count_carry_over_small_train
count_no_carry_over_small_test = count_small_test - count_carry_over_small_test
count_carry_over_large_test = total_count_carry_over_large - count_carry_over_large_train
total_count_no_carry_over_large = total_count_large - total_count_carry_over_large
count_no_carry_over_large_train = count_large_train - count_carry_over_large_train
count_no_carry_over_large_test = count_large_test - count_carry_over_large_test

model_regex = r"Loaded trained_model_(\d{4}_\d{2}_\d{2}_\d{2}_\d{2}_\d{2}).json"
type_file_regex = r"Tests_(A|R)P_([\d\.]+).txt"
 
correct_predictions_regex = r", (\d+) predictions correct."
correct_predictions_train_regex = r", (\d+) train predictions correct."
correct_carry_over_predictions_regex = r", (\d+) carry-over predictions correct."
correct_carry_over_predictions_train_regex = r", (\d+) carry-over train predictions correct."
correct_predictions_small_regex = r", (\d+) small predictions correct."
correct_predictions_small_train_regex = r", (\d+) small train predictions correct."
correct_predictions_carry_over_small_regex = r", (\d+) carry-over small predictions correct."
correct_predictions_carry_over_small_train_regex = r", (\d+) carry-over small train predictions correct"
correct_predictions_large_regex = r", (\d+) large predictions correct."
correct_predictions_large_train_regex = r", (\d+) large train predictions correct."
correct_predictions_carry_over_large_regex = r", (\d+) carry-over large predictions correct."
correct_predictions_carry_over_large_train_regex = r", (\d+) carry-over large train predictions correct"

data = []

for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        match = re.search(type_file_regex, filename)        
        if match:
            epsilon = match.group(2)
            
            with open(os.path.join(folder_path, filename), 'r') as file:
                for line in file:
                    model_match = re.search(model_regex, line)
                    if model_match:
                        current_time = model_match.group(1)
                        continue

                    predictions = re.search(correct_predictions_regex, line)
                    if predictions:
                        total_correct_predictions = int(predictions.group(1))
                        continue  
                        
                    predictions_train = re.search(correct_predictions_train_regex, line)
                    if predictions_train:
                        correct_predictions_train = int(predictions_train.group(1))
                        continue                     

                    carry_over_predictions = re.search(correct_carry_over_predictions_regex, line)
                    if carry_over_predictions:
                        total_correct_carry_over_predictions = int(carry_over_predictions.group(1))
                        continue 

                    carry_over_predictions_train = re.search(correct_carry_over_predictions_train_regex, line)
                    if carry_over_predictions_train:
                        correct_carry_over_predictions_train = int(carry_over_predictions_train.group(1))
                        continue 

                    predictions_small = re.search(correct_predictions_small_regex, line)
                    if predictions_small:
                        total_correct_predictions_small = int(predictions_small.group(1))
                        continue 

                    predictions_small_train = re.search(correct_predictions_small_train_regex, line)
                    if predictions_small_train:
                        correct_predictions_small_train = int(predictions_small_train.group(1))
                        continue                     

                    predictions_carry_over_small = re.search(correct_predictions_carry_over_small_regex, line)
                    if predictions_carry_over_small:
                        total_correct_predictions_carry_over_small = int(predictions_carry_over_small.group(1))
                        continue 

                    predictions_carry_over_small_train = re.search(correct_predictions_carry_over_small_train_regex, line)
                    if predictions_carry_over_small_train:
                        correct_predictions_carry_over_small_train = int(predictions_carry_over_small_train.group(1))     
                        continue

                    predictions_large = re.search(correct_predictions_large_regex, line)
                    if predictions_large:
                        total_correct_predictions_large = int(predictions_large.group(1))
                        continue 

                    predictions_large_train = re.search(correct_predictions_large_train_regex, line)
                    if predictions_large_train:
                        correct_predictions_large_train = int(predictions_large_train.group(1))
                        continue                     

                    predictions_carry_over_large = re.search(correct_predictions_carry_over_large_regex, line)
                    if predictions_carry_over_large:
                        total_correct_predictions_carry_over_large = int(predictions_carry_over_large.group(1))
                        continue 

                    predictions_carry_over_large_train = re.search(correct_predictions_carry_over_large_train_regex, line)
                    if predictions_carry_over_large_train:
                        correct_predictions_carry_over_large_train = int(predictions_carry_over_large_train.group(1))     
                        analyzed_file_path = f"{analyzed_data_folder}/all_results-{param_type}_{epsilon}.csv"
                        
                        if os.path.exists(analyzed_file_path):
                            analyzed_df = pd.read_csv(analyzed_file_path, delimiter=';', decimal=',')
                            
                            model_data = analyzed_df[(analyzed_df['model'] == current_time)]
                            total_trainings = len(model_data) * 10
                        else:
                            total_trainings = 0

                        correct_predictions_test = total_correct_predictions - correct_predictions_train
                        correct_carry_over_predictions_test = total_correct_carry_over_predictions - correct_carry_over_predictions_train
                        total_correct_no_carry_over_predictions = total_correct_predictions - total_correct_carry_over_predictions
                        correct_no_carry_over_predictions_train = correct_predictions_train - correct_carry_over_predictions_train
                        correct_no_carry_over_predictions_test = correct_predictions_test - correct_carry_over_predictions_test
                        
                        correct_predictions_small_test = total_correct_predictions_small - correct_predictions_small_train
                        correct_predictions_large_test = total_correct_predictions_large - correct_predictions_large_train
                        
                        correct_predictions_carry_over_small_test = total_correct_predictions_carry_over_small - correct_predictions_carry_over_small_train 
                        total_correct_predictions_no_carry_over_small = total_correct_predictions_small - total_correct_predictions_carry_over_small
                        correct_predictions_no_carry_over_small_train = correct_predictions_small_train - correct_predictions_carry_over_small_train
                        correct_predictions_no_carry_over_small_test = correct_predictions_small_test - correct_predictions_carry_over_small_test
                        correct_predictions_carry_over_large_test = total_correct_predictions_carry_over_large - correct_predictions_carry_over_large_train 
                        total_correct_predictions_no_carry_over_large = total_correct_predictions_large - total_correct_predictions_carry_over_large
                        correct_predictions_no_carry_over_large_train = correct_predictions_large_train - correct_predictions_carry_over_large_train
                        correct_predictions_no_carry_over_large_test = correct_predictions_large_test - correct_predictions_carry_over_large_test
                        
                        data.append([current_time, epsilon, total_trainings,
                                     total_count, total_correct_predictions,
                                     count_train, correct_predictions_train,
                                     count_test, correct_predictions_test,
                                     total_count_carry_over, total_correct_carry_over_predictions,
                                     count_carry_over_train, correct_carry_over_predictions_train,
                                     count_carry_over_test, correct_carry_over_predictions_test,
                                     total_count_no_carry_over, total_correct_no_carry_over_predictions,
                                     count_no_carry_over_train, correct_no_carry_over_predictions_train,
                                     count_no_carry_over_test, correct_no_carry_over_predictions_test,
                                     total_count_small, total_correct_predictions_small,
                                     count_small_train, correct_predictions_small_train,
                                     count_small_test, correct_predictions_small_test,
                                     total_count_large, total_correct_predictions_large,
                                     count_large_train, correct_predictions_large_train,
                                     count_large_test, correct_predictions_large_test,
                                     total_count_carry_over_small, total_correct_predictions_carry_over_small,
                                     count_carry_over_small_train, correct_predictions_carry_over_small_train,
                                     count_carry_over_small_test, correct_predictions_carry_over_small_test,
                                     total_count_no_carry_over_small, total_correct_predictions_no_carry_over_small,
                                     count_no_carry_over_small_train, correct_predictions_no_carry_over_small_train,
                                     count_no_carry_over_small_test, correct_predictions_no_carry_over_small_test,
                                     total_count_carry_over_large, total_correct_predictions_carry_over_large,
                                     count_carry_over_large_train, correct_predictions_carry_over_large_train,
                                     count_carry_over_large_test, correct_predictions_carry_over_large_test,
                                     total_count_no_carry_over_large, total_correct_predictions_no_carry_over_large,
                                     count_no_carry_over_large_train, correct_predictions_no_carry_over_large_train,
                                     count_no_carry_over_large_test, correct_predictions_no_carry_over_large_test
                                     ]) 

df = pd.DataFrame(data, columns=[
    'current_time', 'epsilon', 'total_trainings',
    'total_count', 'total_correct_predictions',
    'count_train', 'correct_predictions_train',
    'count_test', 'correct_predictions_test',
    'total_count_carry_over', 'total_correct_carry_over_predictions',
    'count_carry_over_train', 'correct_carry_over_predictions_train',
    'count_carry_over_test', 'correct_carry_over_predictions_test',
    'total_count_no_carry_over', 'total_correct_no_carry_over_predictions',
    'count_no_carry_over_train', 'correct_no_carry_over_predictions_train',
    'count_no_carry_over_test', 'correct_no_carry_over_predictions_test',
    'total_count_small', 'total_correct_predictions_small',
    'count_small_train', 'correct_predictions_small_train',
    'count_small_test', 'correct_predictions_small_test',
    'total_count_large', 'total_correct_predictions_large',
    'count_large_train', 'correct_predictions_large_train',
    'count_large_test', 'correct_predictions_large_test',
    'total_count_carry_over_small', 'total_correct_predictions_carry_over_small',
    'count_carry_over_small_train', 'correct_predictions_carry_over_small_train',
    'count_carry_over_small_test', 'correct_predictions_carry_over_small_test',
    'total_count_no_carry_over_small', 'total_correct_predictions_no_carry_over_small',
    'count_no_carry_over_small_train', 'correct_predictions_no_carry_over_small_train',
    'count_no_carry_over_small_test', 'correct_predictions_no_carry_over_small_test',
    'total_count_carry_over_large', 'total_correct_predictions_carry_over_large',
    'count_carry_over_large_train', 'correct_predictions_carry_over_large_train',
    'count_carry_over_large_test', 'correct_predictions_carry_over_large_test',
    'total_count_no_carry_over_large', 'total_correct_predictions_no_carry_over_large',
    'count_no_carry_over_large_train', 'correct_predictions_no_carry_over_large_train',
    'count_no_carry_over_large_test', 'correct_predictions_no_carry_over_large_test'
])

df.to_excel(output_path, index=False)
print(f"File {output_path} created.")
```
