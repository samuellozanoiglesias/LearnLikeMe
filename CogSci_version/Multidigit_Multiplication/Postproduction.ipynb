```python
import pandas as pd
import numpy as np
from numpy import savetxt

import matplotlib.pyplot as plt
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'

import seaborn as sn
```

```python
contar_pre=np.zeros((3,3))
contar_pre[0,0]=2
contar_pre[0,1]=34
contar_pre[0,2]=72
contar_pre[1,0]=2
contar_pre[1,1]=49
contar_pre[1,2]=87
contar_pre[2,0]=97
contar_pre[2,1]=93
contar_pre[2,2]=100

# Normalizar la matriz contar_pre
contar = np.zeros((3, 3))
contar[0,0]=2
contar[0,1]=34
contar[0,2]=72
contar[1,0]=2
contar[1,1]=49
contar[1,2]=87
contar[2,0]=97
contar[2,1]=93
contar[2,2]=100

# Crear la matriz de confusión sin normalizar
array_datos = contar_pre  # Transponer la matriz
df_cm_datos = pd.DataFrame(array_datos, index=["", "", ""], columns=["", "", ""])


# Crear la matriz de confusión normalizada
array_colores = contar  # Transponer la matriz
df_cm_colores = pd.DataFrame(array_colores.T, index=["", "", ""], columns=["", "", ""])


annotations = np.empty_like(df_cm_datos).astype(str)
for i in range(df_cm_datos.shape[0]):
    for j in range(df_cm_datos.shape[1]):
        annotations[i, j] = f"{int(df_cm_datos.iloc[i, j])}%"  # Agregar el símbolo de porcentaje

# Crear el heatmap con las anotaciones
plt.figure(figsize=(10, 6))
sn.heatmap(df_cm_datos, annot=annotations, cmap="OrRd", fmt='', cbar=True, annot_kws={"size": 24})

plt.show()
```

```python
x = np.arange(1, 10)  # 9 elementos en el eje X
y1 = np.array([2, 35, 72, 2, 49, 87, 98, 93, 100])
y2 = np.array([1, 33, 70, 1, 47, 87, 93, 93, 98])

# Crear el diagrama de barras
fig, ax = plt.subplots(figsize=(40, 16))

# Barras
bar_width = 0.35  # Ancho de las barras
ax.bar(x - bar_width / 2 -0.01, y1, bar_width, color=(233/255, 113/255, 50/255), label='Without carry-over')  # Barra gris
ax.bar(x + bar_width / 2 +0.01, y2, bar_width, color=(192/255, 0/255, 0/255), label='With carry-over')  # Barra roja

# Configuración de los ejes
ax.set_xticks(x)  # Los 9 elementos en el eje X
ax.set_xticklabels([])  # Sin texto en los ticks del eje X
ax.set_ylim(0, 100)  # Eje Y de 0 a 100
ax.set_ylabel('Accuracy', fontsize=80)  # Etiqueta en el eje Y más grande

# Aumentar el tamaño de los yticks
ax.tick_params(axis='y', labelsize=50)  # Aumentar tamaño de los yticks
ax.legend(fontsize=60) 
# Eliminar la leyenda
# ax.legend()  # No necesitamos leyenda, así que lo hemos comentado

# Mostrar el gráfico
plt.xticks(rotation=45)  # Rotar las etiquetas del eje X si es necesario
plt.tight_layout()
plt.show()
```
