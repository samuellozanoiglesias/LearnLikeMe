```python
import pandas as pd
import numpy as np
from numpy import savetxt

import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'

import seaborn as sns
```

```python
contar_pre=np.zeros((3,3))
contar_pre[0,0]=2
contar_pre[0,1]=34
contar_pre[0,2]=72
contar_pre[1,0]=2
contar_pre[1,1]=49
contar_pre[1,2]=87
contar_pre[2,0]=97
contar_pre[2,1]=93
contar_pre[2,2]=100

# Normalizar la matriz contar_pre
contar = np.zeros((3, 3))
contar[0,0]=2
contar[0,1]=34
contar[0,2]=72
contar[1,0]=2
contar[1,1]=49
contar[1,2]=87
contar[2,0]=97
contar[2,1]=93
contar[2,2]=100

# Crear la matriz de confusión sin normalizar
array_datos = contar_pre  # Transponer la matriz
df_cm_datos = pd.DataFrame(array_datos, index=["", "", ""], columns=["", "", ""])


# Crear la matriz de confusión normalizada
array_colores = contar  # Transponer la matriz
df_cm_colores = pd.DataFrame(array_colores.T, index=["", "", ""], columns=["", "", ""])


annotations = np.empty_like(df_cm_datos).astype(str)
for i in range(df_cm_datos.shape[0]):
    for j in range(df_cm_datos.shape[1]):
        annotations[i, j] = f"{int(df_cm_datos.iloc[i, j])}%"  # Agregar el símbolo de porcentaje

# Crear el heatmap con las anotaciones
plt.figure(figsize=(10, 6))
sns.heatmap(df_cm_datos, annot=annotations, cmap="OrRd", fmt='', cbar=True, annot_kws={"size": 24})

plt.show()
```

```python
y_no_carry = np.array([30.7, 33.4, 46.6, 47.6])
y_carry = np.array([36.1, 40.7, 46.5, 48.4])
x = np.arange(1, y_carry.shape[0] + 1)

# Crear el diagrama de barras
fig, ax = plt.subplots(figsize=(25, 15))

# Barras
bar_width = 0.35  # Ancho de las barras
ax.bar(x - bar_width / 2 -0.01, y_no_carry, bar_width, color=(233/255, 113/255, 50/255), label='Without carry-over')  # Barra gris
ax.bar(x + bar_width / 2 +0.01, y_carry, bar_width, color=(192/255, 0/255, 0/255), label='With carry-over')  # Barra roja

# Configuración de los ejes
ax.set_xticks(x)  # Los 9 elementos en el eje X
ax.set_xticklabels([])  # Sin texto en los ticks del eje X
ax.set_ylim(0, 100)  # Eje Y de 0 a 100
ax.set_ylabel('Average Errors %', fontsize=80)  # Etiqueta en el eje Y más grande

# Aumentar el tamaño de los yticks
ax.tick_params(axis='y', labelsize=50)  # Aumentar tamaño de los yticks
ax.legend(fontsize=60) 
# Eliminar la leyenda
# ax.legend()  # No necesitamos leyenda, así que lo hemos comentado

# Mostrar el gráfico
plt.xticks(rotation=45)  # Rotar las etiquetas del eje X si es necesario
plt.tight_layout()
plt.show()
```

```python
# Valores de y
y_small_no_carry = 16.5
y_small_carry = 27.3
y_large_no_carry = 39.9
y_large_carry = 48.5

# Barras de error
error_small_no_carry = 6.2
error_small_carry = 9.4
error_large_no_carry = 11.5
error_large_carry = 12.7

# Valores de x (debe haber 4 posiciones en total)
x = np.arange(1, 3)

# Colores a utilizar
color_1 = (135/255, 206/255, 235/255)  # SMALL (azul claro)
color_2 = (0/255, 0/255, 139/255)      # LARGE (azul oscuro)
color_3 = (233/255, 113/255, 50/255)   # NO CARRY (naranja)
color_4 = (192/255, 0/255, 0/255)      # CARRY (rojo)

# Configurar la apariencia de las franjas
mpl.rcParams['hatch.linewidth'] = 10  # Ancho de las franjas

# Crear el diagrama de barras
fig, ax = plt.subplots(figsize=(25, 15))

# Configuración para las barras de error
error_kw = dict(elinewidth=3, ecolor='black', capsize=15)

# Barras
bar_width = 0.35  # Ancho de las barras

# Barra 1 (sin acarreo pequeña)
ax.bar(x[0] - bar_width / 2 - 0.01, y_small_no_carry, bar_width, 
       color=color_3, label='Without carry-over', yerr=error_small_no_carry, error_kw=error_kw)

# Barra 2 (con acarreo pequeña)
ax.bar(x[0] + bar_width / 2 + 0.01, y_small_carry, bar_width, 
       color=color_4, label='With carry-over', yerr=error_small_carry, error_kw=error_kw)

# Barra 3 (sin acarreo grande)
ax.bar(x[1] - bar_width / 2 - 0.01, y_large_no_carry, bar_width, 
       color=color_3, yerr=error_large_no_carry, error_kw=error_kw)

# Barra 4 (con acarreo grande)
ax.bar(x[1] + bar_width / 2 + 0.01, y_large_carry, bar_width, 
       color=color_4, yerr=error_large_carry, error_kw=error_kw)

# Configuración de los ejes
ax.set_xticks(x)  # Los 4 elementos en el eje X
ax.set_xticklabels(["Small", "Large"], fontsize=70)  # Etiquetas del eje X
ax.set_ylim(0, 100)  # Eje Y de 0 a 100
ax.set_ylabel('Average Error (%)', fontsize=80)  # Etiqueta en el eje Y más grande
ax.set_xlabel('Problem size', fontsize=80)  # Etiqueta en el eje Y más grande

# Aumentar el tamaño de los yticks
ax.tick_params(axis='x', labelsize=80, length=25, width=1)  # Ticks más largos y gruesos en X
ax.tick_params(axis='y', labelsize=50, length=10, width=1)  # Aumentar tamaño de los yticks
ax.legend(fontsize=80)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
